```python
# Step 12: Create a new column with column name as Product Color which displays the phone color
```

```python
# Create a function to extract color and extract it using regular expression
def extract_color(product_name):
    match = re.search(r'\(([^,]+),', product_name) # Regular expression to match the color name
    return match.group(1).strip() if match else None

# Apply the function to create the 'product color' column
df['Product Color'] = df['Product Name'].apply(extract_color)
df.head(20)
```

```python
# Step 13: Find all the unique colors from the dataset.
```

```python
unique_colors = df['Product Color'].unique()
print(unique_colors)
```

```python
# Step 14: Find and display in new column whether the phone comes with accessories such as power adaptor, earpods etc.
```

```python
def check_accessories(Product_Name):
    keyword = "Includes EarPods, Power Adapter"
    if keyword in Product_Name:
        return "Yes"
    return "No"

df['Other Accessories'] = df['Product Name'].apply(check_accessories)
df.head(15)
```

```python
# Step 15: Sort the most discounted to least discounted phones
```

```python
df.sort_values(by=['Discount Percentage'], ascending = False)
```

```python
# Step 16 : Add a new column with product model number
```

```python
df['Model Name'] = df['Product Name'].str[6:15]
df.head()
```

```python
# Step 17: Fill the missing values with mean MRP of specific model number.
```

```python
# Calculate mean price of each model.
# transform('mean') ensures that the mean price for each Model Name is broadcasted across all rows that belong to that Model Name.
mean_prices = df.groupby('Model Name')['Mrp'].transform('mean')


# fill the missing values with mean values using fillna()
df['Mrp']= df['Mrp'].fillna(mean_prices)
# Round the decimals to 2 
df['Mrp']= df['Mrp'].round(2)
# Check the missing value count
print(df.isnull().sum())
df.tail(10)

```

```python
# Step 18: Aggregate the count of phones by model number
```

```python
df.groupby('Model Name').size().reset_index(name='Count').sort_values(by='Count', ascending=True)
```

```python
# Step 19 : Plot a scatter plot showcasing each product with amount of discount they are providing in the sale.
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Model Name'], df['Discount Percentage'], color='blue', alpha=0.7)
plt.title('iPhone Sale Model Name vs Discount')
plt.xlabel('Model Name')
plt.ylabel('Discount (%)')
plt.grid()
plt.show()
```

```python
# Step 20: Plot a bar graph for number of reviews per each iphone model.
```

```python
review_counts = df.groupby('Model Name')['Number Of Reviews'].count()

# Step 2: Plot the bar chart
plt.figure(figsize=(10, 8))
colors = ['skyblue', 'salmon', 'lightgreen', 'orange', 'purple', 'lightcoral']
review_counts.plot(kind='bar', color= colors[:len(review_counts)])

plt.title('Number of Reviews Per iPhone Model.')
plt.xlabel('Model Name')
plt.ylabel('Number of Reviews (in thousands).')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()  # Adjusts layout to prevent clipping of labels
plt.show()
```
